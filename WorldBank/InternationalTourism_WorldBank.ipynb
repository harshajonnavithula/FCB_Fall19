```python
internationaltourismdataSlope=internationaltourismdata.sort_values('slope', ascending=False)
internationaltourismdataPercent=internationaltourismdata.sort_values('percent', ascending=False)
```

```python
top5Slope = internationaltourismdataSlope.iloc[0:5, 0:9]
top5tSlope = top5Slope.T
top5Percent = internationaltourismdataPercent.iloc[0:5, 0:9]
top5tPercent = top5Percent.T
```

```python
top5tdataSlope = top5tSlope[0:8]
top5tdataPercent = top5tPercent[0:8]
```

```python
top5tdataSlope.columns  = top5Slope['Country Name']
top5tdataPercent.columns  = top5Percent['Country Name']
```

```python
ax = top5tdataSlope.plot()
ax.set_xlabel("Years")
ax.set_ylabel("Tourism expenditures in current USD ($10B) Highest Rate of Change")
ax.set_xticklabels(top5tdataSlope.index, rotation=0)
```

```python
ax = top5tdataPercent.plot()
ax.set_xlabel("Years")
ax.set_ylabel("Tourism expenditures in current USD ($10B) (%) Increase")
ax.set_xticklabels(top5tdataPercent.index, rotation=0)
```
