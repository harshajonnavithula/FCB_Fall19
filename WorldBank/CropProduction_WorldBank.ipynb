```python
import pandas
cropproduction = pandas.read_csv('CropProduction.csv')
cropproductioncountry=cropproduction.iloc[:,0:1]
cropproductiondata = cropproduction.iloc[:, 54:61]
cropproductiondata = cropproductiondata.dropna()
cropproductiondata = cropproductiondata.join(cropproductioncountry)
cropproductiondata.to_csv('CropProductionCLEANED.csv')
```

```python
cropproductiondata['slope'] = (cropproductiondata['2016']-cropproductiondata['2010'])/6
cropproductiondata['percent'] = (cropproductiondata['2016']-cropproductiondata['2010'])/cropproductiondata['2010']
```

```python
cropproductiondata.sort_values('slope', ascending=False)
```

```python
cropproductiondata["2010"].map(float)
cropproductiondata["2011"].map(float)
cropproductiondata["2012"].map(float)
cropproductiondata["2013"].map(float)
cropproductiondata["2014"].map(float)
cropproductiondata["2015"].map(float)
cropproductiondata["2016"].map(float)
cropproductiondataSlope=cropproductiondata.sort_values('slope', ascending=False)
cropproductiondataPercent=cropproductiondata.sort_values('percent', ascending=False)
```

```python
top5Slope = cropproductiondataSlope.iloc[0:5, 0:9]
top5tSlope = top5Slope.T
top5Percent = cropproductiondataPercent.iloc[0:5, 0:9]
top5tPercent = top5Slope.T
```

```python
top5tdataSlope = top5tSlope[0:8]
top5tdataPercent = top5tPercent[0:8]
```

```python
top5tdataSlope.columns  = top5Slope['Country Name']
top5tdataPercent.columns  = top5Percent['Country Name']
```

```python
top5tdataSlope = top5tSlope[0:7]
top5tdataSlope
top5tdataSlope.columns  = top5Slope['Country Name']
top5tdataSlope
```

```python
top5tdataPercent = top5tPercent[0:7]
top5tdataPercent
top5tdataPercent.columns  = top5Percent['Country Name']
top5tdataPercent
```

```python
ax = top5tdataSlope.plot()
ax.set_xlabel("Years")
ax.set_ylabel("Crop Production Index Highest Rate of Change")
ax.set_xticklabels(top5tdataSlope.index, rotation=0)
```

```python
ax = top5tdataPercent.plot()
ax.set_xlabel("Years")
ax.set_ylabel("Crop Production Index (%) Increase")
ax.set_xticklabels(top5tdataPercent.index, rotation=0)
```
